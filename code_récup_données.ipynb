```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import numpy as np
```

### Chargement de la page web

```python
r = requests.get("https://listfist.com/list-of-countries-by-average-temperature")

#Convertir  un objet beautiful soup
soup = bs(r.content)

#Afficher le HTML
contents = soup.prettify()
print(contents)
```

```python
info_box = soup.find("tbody")
print(info_box.prettify())
```

```python
info_rows = info_box.find_all("tr")
for row in info_rows:
    print(row.prettify())
```

```python
pays = {}

for index, row in enumerate(info_rows):
    #if index == 0:
     #   continue
    if index < 196:
        #print(index)
        key = row.select('td')[2].get_text(' ', strip=True)
        value = row.select('td')[3].get_text(' ', strip=True)
        pays[key] = value

print(pays)
```

```python
pays_1 = {}

for index, row in enumerate(info_rows):
    #if index == 0:
     #   continue
    if index < 196:
        #print(index)
        key = row.select('td')[2].get_text(' ', strip=True)
        value = row.select('td')[4].get_text(' ', strip=True)
        pays_1[key] = value
```

```python
pays_2 = {}

for index, row in enumerate(info_rows):
    #if index == 0:
     #   continue
    if index < 196:
        #print(index)
        key = row.select('td')[2].get_text(' ', strip=True)
        value = row.select('td')[5].get_text(' ', strip=True)
        pays_2[key] = value
```

```python
pays_3 = {}
for index, row in enumerate(info_rows):
    #if index == 0:
     #   continue
    if index < 196:
        #print(index)
        key = row.select('td')[2].get_text(' ', strip=True)
        value = row.select('td')[6].get_text(' ', strip=True)
        pays_3[key] = value
```

```python
pays_3
```

```python
test = pd.DataFrame(data=pays, index=[0])
test_1 = pd.DataFrame(data=pays_1, index=[0])
test_2 = pd.DataFrame(data=pays_2, index=[0])
test_3 = pd.DataFrame(data=pays_3, index=[0])
```

```python
df = test.T
df1 = test_1.T
df2 = test_2.T
df3 = test_3.T
df.reset_index(inplace=True)
df.rename(columns={'index':'Pays', 0:'temprature moyenne(C)'}, inplace=True)
df1.reset_index(inplace=True)
df1.rename(columns={'index':'Pays', 0:'tmin'}, inplace=True)
df2.reset_index(inplace=True)
df2.rename(columns={'index':'Pays', 0:'tmax'}, inplace=True)
df3.reset_index(inplace=True)
df3.rename(columns={'index':'Pays', 0:'Variation_temp'}, inplace=True)
```

```python
df3
```

```python
data = df.merge(df1, on='Pays')
data = data.merge(df2, on='Pays')
data = data.merge(df3, on='Pays')
```
